1. Заменим html-сущности (к примеру: &lt; &gt; &amp;). "&lt;" заменим на “<” и "&amp;" заменим на “&”)""". Сделаем это с помощью HTMLParser.unescape(). Всю предобработку делаем в новом столбце 'clean_tweet'

2. Удалим @user из всех твитов с помощью паттерна "@[\w]*". Для этого создадим функцию: 
 - для того, чтобы найти все вхождения паттерна в тексте, необходимо использовать re.findall(pattern, input_txt)
 - для для замены @user на пробел, необходимо использовать re.sub()
при применении функции необходимо использовать np.vectorize(function)


```python
def del_user(x):
    result = re.findall('@[\w]+', x)
    if len(result) > 0:
        return re.sub('@[\w]+', ' ', x)
    else:
        return x

vec_del_user = np.vectorize(del_user)
```

```python
df['clean_tweet'] = vec_del_user(df['clean_tweet'])
df.head()
```

3. Изменим регистр твитов на нижний с помощью .lower()

```python
def func_lower(x):
    return x.lower()

vec_lower = np.vectorize(func_lower)
```

```python
df['clean_tweet'] = vec_lower(df['clean_tweet'])
```

4. Заменим сокращения с апострофами (пример: ain't, can't) на пробел, используя apostrophe_dict. Для этого необходимо сделать функцию: для каждого слова в тексте проверить (for word in text.split()), если слово есть в словаре apostrophe_dict в качестве ключа (сокращенного слова), то заменить ключ на значение (полную версию слова).

```python
def del_apostrophe(tweet):
        
    def _dashrepl(match_key):
        key = match_key.group(0)        
        if key in apostrophe_dict.keys():
            return apostrophe_dict[key]
        else:
            return re.sub(r"'", "", key)
                
    return re.sub(r"(\w+')+\w+", _dashrepl, tweet)

vec_del_apostrophe = np.vectorize(del_apostrophe)
```

```python
df['clean_tweet'] = vec_del_apostrophe(df['clean_tweet'])
df.head()
```

5. Заменим сокращения на их полные формы, используя short_word_dict. Для этого воспользуемся функцией, используемой в предыдущем пункте.

```python
def short_to_full(tweet):
    
    keys = set(short_word_dict.keys()) & set(tweet.split())
    for key in keys:
        tweet = re.sub(r"\b"+key+r"\b", short_word_dict[key], tweet)

    return tweet

vec_short_to_full = np.vectorize(short_to_full)
```

```python
df['clean_tweet'] = vec_short_to_full(df['clean_tweet'])
df.head()
```

6. Заменим эмотиконы (пример: ":)" = "happy") на пробелы, используя emoticon_dict. Для этого воспользуемся функцией, используемой в предыдущем пункте.

```python
def emoticon_to_word(tweet):
    for key in emoticon_dict.keys():
        if key in tweet:
            tweet = tweet.replace(key, emoticon_dict[key])
    tweet = re.sub(r"(\S)(happy|sad)", r"\1 \2", tweet) # если смайлик напечатан после слова без пробела
    return tweet

vec_emoticon_to_word = np.vectorize(emoticon_to_word)
```

```python
df['clean_tweet'] = vec_emoticon_to_word(df['clean_tweet'])
```

7. Заменим пунктуацию на пробелы, используя re.sub() и паттерн r'[^\w\s]'

```python
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: re.sub(r"[^\w\s]", ' ', x))
```

8. Заменим спец. символы на пробелы, используя re.sub() и паттерн r'[^a-zA-Z0-9]'

```python
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: re.sub(r"[^a-zA-Z0-9]", ' ', x))
```

9. Заменим числа на пробелы, используя re.sub() и паттерн r'[^a-zA-Z]'

```python
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: re.sub('[^a-zA-Z]', ' ', x))
```

10. Удалим из текста слова длиной в 1 символ, используя ' '.join([w for w in x.split() if len(w)>1])

```python
def one_remove(x):
    return ' '.join([w for w in x.split() if len(w)>1])

vec_one_remove = np.vectorize(one_remove)
```

```python
df['clean_tweet'] = df['clean_tweet'].apply(lambda x: vec_one_remove(x))
```

11. Поделим твиты на токены с помощью nltk.tokenize.word_tokenize, создав новый столбец 'tweet_token'.

```python
nltk.download('punkt')
```

```python
df['tweet_token'] = df['clean_tweet'].apply(lambda x: nltk.tokenize.word_tokenize(x))
df.head()
```

12. Удалим стоп-слова из токенов, используя nltk.corpus.stopwords. Создадим столбец 'tweet_token_filtered' без стоп-слов.

```python
nltk.download('stopwords')
```

```python
def stopwords_remove(tweet_token):
    if len(tweet_token) > 0:
        return [word for word in tweet_token
                if word not in nltk.corpus.stopwords.words('english')] 
    else:
        return tweet_token
```

```python
df['tweet_token_filtered'] = df['tweet_token'].apply(stopwords_remove)
df.head()
```

13. Применим стемминг к токенам с помощью nltk.stem.PorterStemmer. Создадим столбец 'tweet_stemmed' после применения стемминга.

```python
stemmer = nltk.stem.PorterStemmer()
```

```python
def stem_tweet(tweet_token):
    return [stemmer.stem(word) for word in tweet_token]
```

```python
df['tweet_stemmed'] = df['tweet_token_filtered'].apply(stem_tweet)
df.head()
```

14. Применим лемматизацию к токенам с помощью nltk.stem.wordnet.WordNetLemmatizer. Создадим столбец 'tweet_lemmatized' после применения лемматизации.

```python
nltk.download('wordnet')
```

```python
lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()
```

```python
def lemm_tweet(tweet_token):
    return [lemmatizer.lemmatize(word) for word in tweet_token]
```

```python
df['tweet_lemmatized'] = df['tweet_token_filtered'].apply(lemm_tweet)
df.head()
```
